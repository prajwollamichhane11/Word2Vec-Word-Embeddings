```
completeText = []
with open("./nepali.txt","r",encoding="utf-8") as f:
    completeText.append(f.read())
```

```
stopwordNP = []
with open('./stopwordsNepali.txt','r',encoding='utf-8') as file:
  stopwordNP.append(file.read())
```

```
corpus = re.sub(r'(\n)'," ",stopwordNP[0])
corpus = corpus.strip(" ")
stopwordNP = set(corpus.split(" "))
stopwordNP
```

```
tokenized = re.split(r'\s+',completeText)
filterSW = [word for word in tokenized if not word in stopwordNP]
filterSW[:10]
```

```
model = gensim.models.Word2Vec(
    [filterSW],
    size=100,
    window=10,
    min_count=1)
model.train(completeText,total_examples=len(completeText),epochs=10)
```
